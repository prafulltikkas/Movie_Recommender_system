```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast (ml module)
```

```python
def convert(text):      # created convert function literal eval function used
    L = []
    for i in ast.literal_eval(text): 
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)   # remove null values
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):  # cast actor name convert3 function created
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head() #cast name display
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):  #director name fetch_director function created
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) #director name shown changes in data
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):  #crew director name space removed,cast name space removed
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head() #dispaly data
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] #added this column in movies tags
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew']) # drop columns overview,genres,keywords,cast
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) #in tags added 5columns summary
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer #importing count vectorizer converts text into numerical matrix
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray() 
```

```python
vector.shape  #4806row,5000 column shown
```

```python
from sklearn.metrics.pairwise import cosine_similarity #used for measures non zero vectors
```

```python
similarity = cosine_similarity(vector) #showing non zero vectors
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]  
```

```python
def recommend(movie):      #recommend(movie) function created ,sorted list,loop created title will be display
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
    
```

```python
recommend('Gandhi') #movie name summary will be display
```
